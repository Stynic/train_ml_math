```python
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np
import matplotlib.pyplot as plt
def hi_2(obs:list, exp:list):
    if sum(obs) == sum(exp):
        amount = 0
        for i in range(len(obs)):
            amount += ((obs[i] - exp[i]) / np.sqrt(exp[i]))**2 
    return amount
evarage_hi_2 = [] 

for i in range(10000): 
    expected = (30, 30)
    exp1 = pd.DataFrame({'observed': random.choices(['Орел', 'Решка'], k=60)})
    observed = exp1['observed'].value_counts().values
    evarage_hi_2.append(hi_2(observed, expected))
print(f'Mean: {sum(evarage_hi_2) / len(evarage_hi_2)}')

plt.hist(evarage_hi_2)
plt.show()
```

```python
from scipy.stats import chi2
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))

df_s = 5

x = np.linspace(0, 10, 100)

for i in range(0, df_s):
    df = i + 1
    plt.plot(x, chi2.pdf(x, df), label=f'k={df}')
plt.legend(bbox_to_anchor=(1, 1), fontsize=14)
plt.title('Сhi-square distribution depends on degrees of freedom', fontsize=16)
plt.grid()
plt.show()
```
