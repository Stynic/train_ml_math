```python
n = 81
count = 0

for pair in list(itertools.combinations_with_replacement(range(1, 10), 2)):
    # print(f"{pair}=",sum(pair))
    if sum(pair) == 15:
        # print('-------')
        count += 1
        
print(count)
# respose 2 * 2 учитываем еще противоположные
```
