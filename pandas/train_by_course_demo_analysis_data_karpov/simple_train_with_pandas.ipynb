```python
data_students = pd.read_csv('StudentsPerformance.csv')
data_students.head()
```

```python
query_free = data_students['lunch'] == 'free/reduced'
lunch_true = data_students[query_free].shape[0]
print('share:', lunch_true / data_students.shape[0])
```

```python
data_students.columns
```

```python
data_students.columns = ['gender', 'race/ethnicity', 'parental_level_of_education', 'lunch',
       'test_preparation_course', 'math_score', 'reading_score',
       'writing_score']
```

```python
score = 90
lunch_true = data_students.query('(lunch == "free/reduced") & (math_score>@score)')
lunch_true
```

```python
lunch_true = data_students.query('lunch == "free/reduced"').shape[0]
print('share:', lunch_true / data_students.shape[0])
```

```python
# how to find out all markers in feature
data_students['lunch'].unique()
```

```python
data_students.groupby('lunch').agg({'math score': ['mean', 'std'],
                                    'reading score': ['mean', 'std'],
                                    'writing score': ['mean', 'std']})
```
