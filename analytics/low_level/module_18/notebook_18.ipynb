```python
def get_weater(day, month, year, city):
    host = 'https://www.metaweather.com/api/location'

    def check_date(date):
        return date if date >= 10 else f'0{date}'

    day, month = check_date(day), check_date(month)

    with req.urlopen(f'{host}/search/?query={city}') as session:
        responce = session.read().decode()

    id_city = json.loads(responce)[0]['woeid']
    df = pd.read_json(f'{host}/{id_city}/{year}/{month}/{day}/')
    return df[df['applicable_date'] == f'{year}-{month}-{day}']['the_temp'].mean()
```

### lesson 2
   ,        28  2020 .       ,  %city_id%, %year%, %month%, %day%   , ,    .

```python
import pandas as pd


def get_weater(day, month, year, city):
    host = 'https://www.metaweather.com/api/location'

    def check_date(date):
        return date if date >= 10 else f'0{date}'

    def get_city(city: str) -> int:
        with req.urlopen(f"{host}/search/?query={city}") as session:
            responce = session.read().decode()
        return json.loads(responce)[0]['woeid']

    day, month = check_date(day), check_date(month)
    id_city = get_city('Moscow')
    df = pd.read_json(f'{host}/{id_city}/{year}/{month}/{day}/')
    return df[df['applicable_date'] == f'{year}-{month}-{day}']['the_temp']
```

```python
get_weater(day=28,
           month=4,
           year=2020,
           city='Moscow').to_json()
```
