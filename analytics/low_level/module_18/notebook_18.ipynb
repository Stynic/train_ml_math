### Reading data from Web-service

```python
import urllib.request as req
import json
import pandas as pd
```

```python
url = "https://www.metaweather.com/api/location/search/?query=moscow"
url
```

```python
session = req.urlopen(url)
# will get session with server by adress from url
response = session.read().decode()
# will get responce from the server by request
print(response)
session.close()
# This is for a close session in order not to pour out data or won't get problems with the browser.
```

```python
# use context manager in order to not forget to close session
with req.urlopen(url) as session:
    response = session.read().decode()
    print(response)
```

```python
[data]= json.loads(response)
print(type(data), data, sep='\n\n')
```

```python
id_city = data['woeid']
id_city
```

```python
url_date = f"https://www.metaweather.com/api/location/{id_city}/2022/02/09"
with req.urlopen(url_date) as session:
    responce = session.read().decode()
data = json.loads(responce)
data
```
