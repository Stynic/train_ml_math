### Задание 1.

Есть файл с протоколом регистраций пользователей на сайте (registrations.txt).<br>
Каждая строка содержит информацию о имени, электронной почте и возрасте человека. <br><br>


Надо проверить данные из файла, для каждой строки:
 - присутсвуют все три поля
 - поле имени содержит только буквы
 - поле email содержит @ и .
 - поле возраст является числом от 10 до 99<br>

В результате проверки нужно сформировать два файла
 - registrations_good.log для правильных данных, записывать строки как есть
- registrations_bad.log для ошибочных, записывать строку и вид ошибки.<br>

Для валидации строки данных написать метод, который может выкидывать исключения:
 - НЕ присутсвуют все три поля: ValueError
 - поле имени содержит НЕ только буквы: NotNameError (кастомное исключение)
 - поле email НЕ содержит @ и .(точку): NotEmailError (кастомное исключение)
 - поле возраст НЕ является числом от 10 до 99: ValueError
Вызов метода обернуть в try-except.

```python
### YOUR CODE HERE ###
class NotNameError(Exception):
    def __init__(self, message=None, input_data=None):
        self.message = message or \
            'поле имени содержит НЕ только буквы: NotNameError'
        self.input_data = input_data


class NotEmailError(Exception):
    def __init__(self, message=None, input_data=None):
        self.message = message or \
            'поле email НЕ содержит @ и .(точку): NotEmailError'
        self.input_data = input_data


def write_validation_file(line: str):
    validation_file.append(line)


def write_error_file(line: str, error: str, input_data=None):
    line = line.strip()
    if isinstance(error, ValueError):
        print(f'error: {error}')
        if '10 to 99' in error.args[0]:
            error_file.append(
                f'{line}; поле возраст НЕ является числом от 10 до 99: ValueError; entered data: {input_data}')
        else:
            error_file.append(
                f'{line}; НЕ присутсвуют все три поля: ValueError; entered data: {input_data}')
    elif isinstance(error, NotNameError):
        error_file.append(
            f'{line}; поле имени содержит НЕ только буквы: NotNameError; entered data: {input_data} ')
    elif isinstance(error,  NotEmailError):
        error_file.append(
            f'{line}; поле email НЕ содержит @ и .(точку): NotEmailError; entered data: {input_data} ')


def validation_users(line: str):
    try:
        line = line.strip()
        name, email, age = list(
            map(lambda string: string.strip(), line.split(' ')))
        print(f'Worked line: {line}')
        if not name.isalpha():
            raise NotNameError(input_data={'name': name})
        if not ('@' and '.') in email:
            raise NotEmailError(input_data={'email': email})
        if not 10 < int(age) < 99:
            raise ValueError('not int from 10 to 99')

    except ValueError as exc:
        write_error_file(line=line, error=exc, input_data=line)
    except NotNameError as exc:
        print(f'{exc.message} {exc.input_data}')
        write_error_file(line=line, error=exc, input_data=exc.input_data)
    except NotEmailError as exc:
        print(f'{exc.message} {exc.input_data}')
        write_error_file(line=line, error=exc, input_data=exc.input_data)
    else:
        write_validation_file(line=line)


if __name__ == '__main__':
    validation_file = []
    error_file = []

    file = 'registrations_.txt'
    with open(file, 'r') as file:
        for index, line in enumerate(file, 1):
            validation_users(line=line)
    with open('registrations_good.log', 'w+') as file_good:
        file_good.write('\n'.join(validation_file))
    with open('registrations_bad.log', 'w+') as file_bad:
        file_bad.write('\n'.join(error_file))
```
