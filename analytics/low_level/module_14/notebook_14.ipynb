```python
tesla2014 = data.loc['2014']
tesla2014
# filtered out data only for the 2014 year
```

```python
import matplotlib.pyplot as plt
```

```python
price_tesla_share = 250
```

```python
max_cost_share = tesla2014['High'].max()
date_max_cost = tesla2014[lambda x: x['High'] == max_cost_share]
```

```python
plt.figure(figsize=(10, 10))
plt.xlabel('Dates')
plt.ylabel('US$')
plt.title('Share Tesla')


plt.axhline(y=price_tesla_share, ls='--',
            label='bought the share by 250', color='r')

plt.axhline(y=max_cost_share, ls='--',
           label=f'Max cost the share {round(max_cost_share, 1)}',
           color='y')
plt.axvline(x=date_max_cost.index[0], color='y', ls='--')

line_high = plt.plot(tesla2014.index, tesla2014['High'], label='High')
line_low = plt.plot(tesla2014.index, tesla2014['Low'], label='Low')
plt.setp([line_high, line_low], alpha=0.7, linewidth=0.7)

plt.legend(loc='lower right', title='Legend')
plt.show()
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('AB_NYC_2019.csv')
data.head()
```

```python
data.dtypes
# checked type data
```

```python
data.isnull().sum()
# check zero rows 
```

```python
data.drop(['id', 'name', 'host_name', 'last_review'], axis=1, inplace=True)
```

```python
data[data['reviews_per_month'].isna()]['number_of_reviews'].sum()
```

```python
data.fillna({'reviews_per_month': 0}, inplace=True)
```

```python
print(data.neighbourhood_group.unique())
print(data.room_type.unique())
# look at the type data
```

## train function Scatter

```python
plt.figure(figsize=(10, 5))
for name, group in data.groupby('neighbourhood_group'):
    plt.scatter(group['longitude'], group['latitude'], s=0.1, label=name)

plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('Accommodation location by neighbourhood group')
plt.legend(markerscale=10)
plt.show()
```

```python
data.groupby('neighbourhood_group').agg('count')
```

```python
# explicitly choose the color
import matplotlib
colors = matplotlib.colors.get_named_colors_mapping()
```

```python
color_map = {
    'Brooklyn': colors['xkcd:cloudy blue'],

    'Manhattan': colors['xkcd:azul'],

    'Queens': colors['xkcd:electric lime'],

    'Staten Island': colors['xkcd:fresh green'],

    'Bronx': colors['xkcd:warm purple']
}
```

```python
plt.figure(figsize=(10, 5))
for name, group in data.groupby(['neighbourhood_group']):
    plt.scatter(group['longitude'],
                group['latitude'],
                s=0.1,
                c=color_map[name],
                label=name)
```

## Bar, barh

```python
room_type_size = data.groupby('room_type').size()
room_type_size
```

```python
plt.bar(room_type_size.index, room_type_size.values)
plt.xlabel('Room types')
plt.ylabel('Room quatify')
plt.title('Accommodation types in numbers')
plt.show()
```

```python
plt.barh(room_type_size.index, room_type_size.values)
plt.ylabel('Room types')
plt.xlabel('Room quatify')
plt.title('Accommodation types in numbers')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
x = np.asarray([0, 1, 2])
width = 0.15
i = -2
for name, group in data.groupby(['neighbourhood_group']):
    plt.bar(x - i*width, group.groupby('room_type').size().values, width, label=name)
    i += 1
plt.title('Accomodation types in numbers by neighbourhood_group')
plt.xticks(x, list(data.groupby('room_type').groups.keys()))
plt.legend()
plt.show()
```

## Hist 

```python
plt.figure(figsize=(8, 8))
filtered_data = data[data['price'] < 1000]
for name, group in filtered_data.groupby(['neighbourhood_group']):
    plt.hist(group['price'], label=name, bins=50, alpha=0.5)
plt.xlabel('Price, $')
plt.ylabel('Accomodation quantity')
plt.title('Price distribution by neighbourhood_group')
plt.legend()
plt.show()
```

## Pie

```python
neighbourhood_group_size = data.groupby(['neighbourhood_group']).size()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(neighbourhood_group_size.values, labels=neighbourhood_group_size.index, autopct='%1.1f%%')
plt.legend()
plt.title('Accomodation share by neighbourhood_group')
plt.show()
```
