# The module 14 by object analytics : Matplotlib(1)

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

## The Simple plot for trigonometric function

```python
x = np.linspace(-1, 1, 50)
y = np.arcsin(x)
```

```python
plt.plot(x, y)
```

## try to drow one point

```python
x = 5
y = -5
plt.plot(x, y, 'bo')
# b-blue color and o - form of the point 
```

## several lines on the one plot

```python
x = np.linspace(-1, 1, 50)
plt.plot(x, x**2, x, x**3)
```

## The several lines on the plot with the explicit format

```python
x = np.linspace(-1, 1, 50)
plt.plot(x, x**2, 'r*', x, x**3, 'b^')
plt.show()
plt.plot(x, x**2, 'g*', x, x**3, 'y^')
plt.show()
```

## lesson 2

```python
plt.plot(data.index, data['High'], label='High')
plt.plot(data.index, data['Low'], label='Low')
plt.title('Tesla stock prices')
plt.xlabel('Dates')
plt.ylabel('Stock prices')
plt.legend()
plt.show()
```

```python
line_high, = plt.plot(data.index, data['High'])
line_low, = plt.plot(data.index, data['Low'])
plt.title('Tesla stock prices')
plt.xlabel('Dates')
plt.ylabel('Stock prices')
plt.legend([line_high, line_low], ['High', 'Low'])
plt.show()
```

```python
plt.plot(data.index, data['High'], label='High', linewidth=0.5, alpha=0.5)
plt.plot(data.index, data['Low'], label='Low', linewidth=0.5, alpha=0.5)
plt.title('Tesla stock prices')
plt.xlabel('Dates')
plt.ylabel('Stock prices')
plt.legend()
plt.show()
```

```python
first_profit_report_date = pd.Timestamp(day=8, month=5, year=2013)
buy_price = 223
```

```python
plt.figure(figsize=(18, 10))

line_high, = plt.plot(data.index, data['High'], label='High')
line_low, = plt.plot(data.index, data['Low'], label='Low')
plt.setp([line_high, line_low], linewidth=0.5, alpha=0.5)

plt.axvline(x=first_profit_report_date, ls='--',
            color='y', label=first_profit_report_date)
plt.axhline(y=buy_price, ls='--', label=f'Buy price {buy_price}$')

plt.title('Tesla stock prices')
plt.xlabel('Dates')
plt.ylabel('Stock prices, US$')
plt.legend()

plt.grid()
plt.show()
plt.savefig('tesla.png')
```
