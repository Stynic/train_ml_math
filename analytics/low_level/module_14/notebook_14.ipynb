# The module 14 by object analytics : Matplotlib(1)

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

## The Simple plot for trigonometric function

```python
x = np.linspace(-1, 1, 50)
y = np.arcsin(x)
```

```python
plt.plot(x, y)
```

## try to drow one point

```python
x = 5
y = -5
plt.plot(x, y, 'bo')
# b-blue color and o - form of the point 
```

## several lines on the one plot

```python
x = np.linspace(-1, 1, 50)
plt.plot(x, x**2, x, x**3)
```

## The several lines on the plot with the explicit format

```python
x = np.linspace(-1, 1, 50)
plt.plot(x, x**2, 'r*', x, x**3, 'b^')
plt.show()
plt.plot(x, x**2, 'g*', x, x**3, 'y^')
plt.show()
```

## For example on the real data by the cost of the actions Tesla

```python
data = pd.read_csv('Tesla.csv', parse_dates=True, index_col='Date')
data.head()
```

```python
plt.plot(data.index, data['High'])
plt.plot(data.index, data['Low'])
plt.show()
```
