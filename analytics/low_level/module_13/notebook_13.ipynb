```python
import pandas as pd
import numpy as np
df = pd.read_csv('WeightLoss.csv')
```

```python
df.head(10)
```

```python
df['total'] = df['w1'] + df['w2'] + df['w3']
df['total_gr'] = df['total'] * 1000
df.head()
```

```python
df['total_gr'].apply(np.log)
```

```python
df['avloss'] = df.loc[:, 'w1': 'w3'].apply(np.mean, axis=1)
df.head()
```

```python
f = lambda row: row.max() - row.min()
```

```python
df['vrange'] = df.loc[:, 'w1':'w3'].apply(f, axis=1)
df.head(3)
```
