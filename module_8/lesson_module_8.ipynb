## lesson 8.2: find approximetly by function several variables 

```python
from sympy import *
from sympy.plotting import plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
a2, a1, a0 = symbols('a2, a1, a0')
```

```python
MSE = 1/3 * (
    ((a2 *2 + a1 * 200 + a0)- 200)**2 + \
    ((a2 *1 + a1 * 450 + a0)- 200)**2 + \
    ((a2 *3 + a1 * 200 + a0)- 200)**2)

MSE
```

```python
MSE.subs({a2: 100, a1: -1, a0: -100})
```

```python
MSE.subs({a2: -200, a1: 0.1, a0: 5})
```
