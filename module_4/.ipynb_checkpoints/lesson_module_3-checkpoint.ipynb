### Интерактивный пример линейной функции:

```python
%matplotlib inline
from ipywidgets import interactive
from mpl_toolkits.axes_grid1 import AxesGrid
from mpl_toolkits.axes_grid.axislines import SubplotZero
import matplotlib.pyplot as plt
import numpy as np

def line_func(k, b):
    fig = plt.figure(2)
    ax = SubplotZero(fig, 111)
    fig.add_subplot(ax)
    ax.axis['xzero'].set_visible(True)
    ax.axis['yzero'].set_visible(True)
    
    x = np.linspace(-10, 10, num=1000)
    ax.plot(x, k * x + b)
    plt.ylim(-5, 5)
    plt.title('kx + b')
    plt.axis(True)
    plt.grid()
    plt.show()

interactive_plot = interactive(line_func, k=(-5, 5, 0.1), b=(-5, 5, 0.1))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```

### Интерактивный пример квадратичной функции:

```python
%matplotlib inline
from ipywidgets import interactive
from mpl_toolkits.axes_grid1 import AxesGrid
from mpl_toolkits.axes_grid.axislines import SubplotZero
import matplotlib.pyplot as plt
import numpy as np

def square_func(a=1, b=1, c=1):
    fig = plt.figure(2)
    ax = SubplotZero(fig, 111)
    fig.add_subplot(ax)
    ax.axis['xzero'].set_visible(True)
    ax.axis['yzero'].set_visible(True)
    
    x = np.linspace(-4, 4, num=1000)
    x_2 = np.linspace(-4, 4, num=10)
    ax.plot(x, a*x**2 + b*x + c)
    plt.ylim(-4, 4)
    plt.title('a*x**2 + b*x + c')
    plt.axis(True)
    plt.grid()
    plt.show()

interactive_plot = interactive(square_func, a=(-4.0, 4.0, 1),b=(-4.0, 4.0, 0.25), c=(-30, 30, 1))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
