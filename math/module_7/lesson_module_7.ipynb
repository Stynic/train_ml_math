## Do homework 7.4 - find zero points for function with more coefficients

```python
f2 = x**2 + y**2 - 1
f2
```

```python
solve(f2)
```

```python
f2_x = solve(f2)[0][x]
f2_x
```

```python
f2_x.subs(y, 1)
```

```python
plot(f2_x)
```

```python
plot3d(f2)
```
