```python
from sympy import *

a2, a1, a0 = symbols('a2, a1, a0')

MSE = 1/4 * (
(1*a1 + 2*a2 - 5)**2 + \
(5*a1 + 3*a2 - 6)**2 + \
(2*a1 + 4*a2 - 10)**2 + \
(3*a1 + 7*a2 - 8)**2
)
MSE
```

```python
X = Matrix([[1, 2], [5, 3], [2, 4], [3, 7]])
X
```

```python
b = Matrix([5, 6, 10, 8])
b
```

```python
r = ((X.T * X)**(-1) * X.T) * b
r
```

```python
MSE.subs({a1:r[0], a2:r[1]})
```
